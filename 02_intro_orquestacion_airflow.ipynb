# Introduccion: orquestacion y Airflow

Este notebook introduce los conceptos minimos para entender que es un DAG
y el tipo de problemas que resuelve la orquestacion.

La instalacion y el uso practico de Airflow se cubren mas adelante.

## Que es un DAG

- Grafo aciclico dirigido: tareas conectadas por dependencias.
- Sin ciclos: evita ejecuciones infinitas.
- Declarativo: define el flujo, no el scheduler.

## Problemas que resuelve la orquestacion

- Dependencias entre tareas y orden de ejecucion.
- Reintentos, fallas y recuperacion controlada.
- Ventanas de ejecucion y backfills.
- SLA y monitoreo operativo.

## Conceptos de ejecucion

- Scheduling: definir cuando corre un flujo.
- Retries: politicas de reintento por tarea.
- SLA: tiempos maximos aceptables.
- Backfill: ejecuciones historicas controladas.

## Buenas practicas de DAGs

- Idempotencia: una tarea puede repetirse sin efectos colaterales.
- Tareas pequenas y observables.
- Logs claros y metricas de ejecucion.
- Separar definicion del DAG y logica de negocio.
