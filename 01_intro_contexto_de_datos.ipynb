## Ciclo de vida de los datos

- Ingesta: batch y streaming.
- Almacenamiento: data lake, data warehouse.
- Procesamiento: ETL, ELT y EL.
- Consumo: analitica, ML y productos de datos.

## Batch vs streaming

- Batch: procesos por lotes con ventanas definidas.
- Streaming: eventos continuos y ventanas deslizantes.
- La eleccion impacta latencia, costos y orquestacion.
